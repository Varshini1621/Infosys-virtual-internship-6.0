- Q1. Identify columns with missing values and demonstrate at least two methods for imputing or filling these missing values (e.g., mean for numerics, mode for categoricals).

```python

import pandas as pd
import numpy as np
np.random.seed(123)
num1 = np.random.randn(500)
num2 = np.random.uniform(10, 50, size=500)
num3 = np.random.randint(1, 6, size=500).astype('float')
num1[np.random.choice(500, 60, replace=False)] = np.nan
num2[np.random.choice(500, 60, replace=False)] = np.nan
num3[np.random.choice(500, 60, replace=False)] = np.nan
cat1 = np.random.choice(['apple', 'banana', 'grape', np.nan], size=500, p=[0.3, 0.3, 0.3, 0.1])
cat2 = np.random.choice(['A', 'B', 'C', 'D'], size=500)
cat3 = np.random.choice(['Yes', 'No', np.nan], size=500, p=[0.45, 0.45, 0.10])
df_untidy = pd.DataFrame({
    'Score': num1,
    'Height_cm': num2,
    'Rating': num3,
    'Fruit': cat1,
    'Group': cat2,
    'IsActive': cat3
})

df_untidy.loc[df_untidy.sample(frac=0.15, random_state=1).index, 'Height_cm'] = \
    df_untidy['Height_cm'].dropna().astype(str) + 'cm'
df_untidy.loc[df_untidy.sample(frac=0.15, random_state=2).index, 'Rating'] = \
    'Rating: ' + df_untidy['Rating'].dropna().astype(str)
print("Missing values before imputation:\n", df_untidy.isnull().sum())
df_q1 = df_untidy.copy()
df_q1['Rating'] = pd.to_numeric(
    df_q1['Rating'].astype(str).str.replace("Rating: ", ""), errors='coerce'
)
df_q1['Height_cm'] = pd.to_numeric(
    df_q1['Height_cm'].astype(str).str.replace("cm", ""), errors='coerce'
)
df_q1['Score'].fillna(df_q1['Score'].mean(), inplace=True)
df_q1['Rating'].fillna(df_q1['Rating'].median(), inplace=True)
df_q1['Height_cm'].fillna(df_q1['Height_cm'].mean(), inplace=True)
df_q1['Fruit'].fillna(df_q1['Fruit'].mode()[0], inplace=True)
df_q1['IsActive'].fillna(df_q1['IsActive'].mode()[0], inplace=True)
print("\nMissing values after imputation:\n", df_q1.isnull().sum())
df_q1.head()
```

- Q2.  Identify columns with non-numeric (categorical) data and convert them into a numeric format using encoding techniques such as one-hot encoding or label encoding.

```python


import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
np.random.seed(123)

num1 = np.random.randn(500)
num2 = np.random.uniform(10, 50, size=500)
num3 = np.random.randint(1, 6, size=500).astype('float')
num1[np.random.choice(500, 60, replace=False)] = np.nan
num2[np.random.choice(500, 60, replace=False)] = np.nan
num3[np.random.choice(500, 60, replace=False)] = np.nan

cat1 = np.random.choice(['apple', 'banana', 'grape', np.nan], size=500, p=[0.3, 0.3, 0.3, 0.1])
cat2 = np.random.choice(['A', 'B', 'C', 'D'], size=500)
cat3 = np.random.choice(['Yes', 'No', np.nan], size=500, p=[0.45, 0.45, 0.10])

df_untidy = pd.DataFrame({
    'Score': num1,
    'Height_cm': num2,
    'Rating': num3,
    'Fruit': cat1,
    'Group': cat2,
    'IsActive': cat3
})
cat_cols = ['Fruit', 'Group', 'IsActive']
print("Categorical Columns:", cat_cols)
df_label = df_untidy.copy()
le = LabelEncoder()
for col in cat_cols:
    df_label[col] = df_label[col].astype(str)
    df_label[col] = le.fit_transform(df_label[col])
print("\nLabel Encoded DataFrame:\n", df_label.head())
df_onehot = pd.get_dummies(df_untidy, columns=cat_cols, drop_first=True)
print("\nOne-Hot Encoded DataFrame:\n", df_onehot.head())
```

- Q3. Detect any columns in the DataFrame that contain mixed data types (such as numbers stored as strings or strings with prefixes). Write code to clean and convert these columns to appropriate, consistent types.

```python


import pandas as pd
import numpy as np

np.random.seed(123)

num1 = np.random.randn(500)
num2 = np.random.uniform(10, 50, size=500)
num3 = np.random.randint(1, 6, size=500).astype('float')
num1[np.random.choice(500, 60, replace=False)] = np.nan
num2[np.random.choice(500, 60, replace=False)] = np.nan
num3[np.random.choice(500, 60, replace=False)] = np.nan

cat1 = np.random.choice(['apple', 'banana', 'grape', np.nan], size=500, p=[0.3, 0.3, 0.3, 0.1])
cat2 = np.random.choice(['A', 'B', 'C', 'D'], size=500)
cat3 = np.random.choice(['Yes', 'No', np.nan], size=500, p=[0.45, 0.45, 0.10])

df_untidy = pd.DataFrame({
    'Score': num1,
    'Height_cm': num2,
    'Rating': num3,
    'Fruit': cat1,
    'Group': cat2,
    'IsActive': cat3
})
df_untidy.loc[df_untidy.sample(frac=0.15, random_state=1).index, 'Height_cm'] = \
    df_untidy['Height_cm'].dropna().astype(str) + 'cm'
df_untidy.loc[df_untidy.sample(frac=0.15, random_state=2).index, 'Rating'] = \
    'Rating: ' + df_untidy['Rating'].dropna().astype(str)
mixed_cols = []
for col in df_untidy.columns:
    types = df_untidy[col].dropna().map(type).nunique()
    if types > 1:
        mixed_cols.append(col)

print("Mixed-type columns detected:", mixed_cols)
df_clean = df_untidy.copy()
df_clean['Height_cm'] = pd.to_numeric(
    df_clean['Height_cm'].astype(str).str.replace("cm", ""), errors='coerce'
)
df_clean['Rating'] = pd.to_numeric(
    df_clean['Rating'].astype(str).str.replace("Rating: ", ""), errors='coerce'
)

print("\nData types after cleaning:\n", df_clean.dtypes)
print("\nCleaned DataFrame head:\n", df_clean.head())
```

- Q4. Apply scaling and/or normalization techniques (such as Min-Max Scaling and Standardization) to the numerical columns to prepare them for downstream machine learning tasks.

```python


import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler


np.random.seed(123)

num1 = np.random.randn(500)
num2 = np.random.uniform(10, 50, size=500)
num3 = np.random.randint(1, 6, size=500).astype('float')

num1[np.random.choice(500, 60, replace=False)] = np.nan
num2[np.random.choice(500, 60, replace=False)] = np.nan
num3[np.random.choice(500, 60, replace=False)] = np.nan

cat1 = np.random.choice(['apple', 'banana', 'grape', np.nan], size=500, p=[0.3, 0.3, 0.3, 0.1])
cat2 = np.random.choice(['A', 'B', 'C', 'D'], size=500)
cat3 = np.random.choice(['Yes', 'No', np.nan], size=500, p=[0.45, 0.45, 0.10])

df = pd.DataFrame({
    'Score': num1,
    'Height_cm': num2,
    'Rating': num3,
    'Fruit': cat1,
    'Group': cat2,
    'IsActive': cat3
})

df.loc[df.sample(frac=0.15, random_state=1).index, 'Height_cm'] = df['Height_cm'].dropna().astype(str) + 'cm'
df.loc[df.sample(frac=0.15, random_state=2).index, 'Rating'] = 'Rating: ' + df['Rating'].dropna().astype(str)

df['Height_cm'] = pd.to_numeric(df['Height_cm'].astype(str).str.replace("cm", ""), errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'].astype(str).str.replace("Rating: ", ""), errors='coerce')
df['Score'].fillna(df['Score'].mean(), inplace=True)
df['Height_cm'].fillna(df['Height_cm'].mean(), inplace=True)
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
num_cols = ['Score', 'Height_cm', 'Rating']
scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = scaler_minmax.fit_transform(df[num_cols])
scaler_std = StandardScaler()
df_standard = df.copy()
df_standard[num_cols] = scaler_std.fit_transform(df[num_cols])
print("Original Numerical Data (first 5 rows):\n", df[num_cols].head())
print("\nAfter Min-Max Scaling:\n", df_minmax[num_cols].head())
print("\nAfter Standardization:\n", df_standard[num_cols].head())
```
